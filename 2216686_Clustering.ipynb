**Task 2.1**

```python
import matplotlib.pyplot as plt
import numpy as np
#load the contents in the Iris_data.npy into a variable
data= np.load('Iris_data.npy')
#load the contents in the Iris_labels.npy into a variable
labels=np.load('Iris_labels.npy')

# print(labels)#labels contains an array of numbers 0-2
# print(data)
x=(data[:,0])#x-axis
y=(data[:,1])#y-axis
print(x)
# print(y)

#use the lris_labels to color the scatter plot
# source : https://jakevdp.github.io/PythonDataScienceHandbook/05.11-k-means.html#:~:text=The%20%22cluster%20center%22%20is%20the,than%20to%20other%20cluster%20centers.
plt.scatter(x, y, c=labels, s=50, cmap='viridis')

#label the axis and plot
plt.title("Iris Data")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

**Task 2.2**

```python
#import the KMeans from sklearn.cluster
from sklearn.cluster import KMeans

#intialize an instance of a scikit-learn KMeans object
#fit kmeans models to Iris_data
#create model using 4 feature dimensions
kmeans = KMeans(n_clusters=3, random_state=0, n_init="auto").fit(data)

#predict cluster mebership of samples within Iris_data
km_predi=kmeans.predict(data)
# print(predictions)

#use the kmeans clusture membership predictions for samples to color the scatter plot
plt.scatter(x, y, c=km_predi, s=50, cmap='viridis')

#get cluster centers and add k cluster centroids of the model to plot
km_centers = kmeans.cluster_centers_
plt.scatter(km_centers[:, 0], km_centers[:, 1], c='red', marker="X",s=100, alpha=1)

#produce a scatter plot from the results of our cluster
plt.title("KMeans Clustering for Iris_data")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

```

**Task 2.3**

```python
# import Gaussian Mixture from sklearn.mixture
from sklearn.mixture import GaussianMixture

# create a Gaussian Mixture model
# fit the model to Iris_data
gm = GaussianMixture(n_components=3, random_state=0).fit(data)

# prediction the cluster membership of samples within Iris_data
gm_predict=gm.predict(data)
# print(gm_predict)

# #use the Guassian Mixture clusture membership predictions for samples to color the scatter plot
plt.scatter(x, y, c=gm_predict, s=50, cmap='viridis')

# get Guassian means for the model
gm_means=gm.means_
# # print(gm_means)
plt.scatter(gm_means[:, 0], gm_means[:, 1],c='red', marker="X",s=100, alpha=1)

#produce a scatter plot from the results of our cluster
plt.title(" Gaussian Mixture Clustering for Iris_data")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

# get the Guassian posterior probabilities
gm_prob=gm.predict_proba(data)
# print(gm_prob)
# create 3 subplots for the components
fig,ax=plt.subplots(1,3,figsize=(12,4))

# loop through the components
for i in range(3):
  for c in range(1):
    # plot the probas
    ax[i].scatter(x,y,c=gm_prob[:,i])
    #set the title for each sub plot
    ax[i].set_title(f'Component {i}')
    #label the x and y axis for subplots
    #source :https://matplotlib.org/stable/gallery/subplots_axes_and_figures/figure_title.html
    ax[i].set_ylabel("Sepal Length")
    ax[i].set_xlabel("Sepal Width")
plt.show()

```
